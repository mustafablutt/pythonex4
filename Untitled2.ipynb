```python
import numpy as np
from scipy.linalg import toeplitz
import time

m=500
n=200

A=np.random.normal(size=(n,m))
B=toeplitz(np.random.random((m,)), np.random.random((m,)))
```

### 9.1: 

```python
print("A+A= ",A+A)
print("AAT= ",np.dot(A,A.T))
print("ATA= ",np.dot(A.T,A))
print("AB= ",np.dot(A,B))
```

### 9.2: 

```python
b = np.random.random(size=(m,1))
x = np.linalg.solve(B,b)
print("Bx=b: ")
print("b: ",b)
print("x: ",x)
```

### 9.3:

```python
print("AF= ",np.linalg.norm(A,ord='fro'))
print("B INFINITY= ",np.linalg.norm(B,np.inf))
e,v=np.linalg.eig(B)
print("Maximum eigen= ", e.max())
print("Minumum eigen= ",e.min())

```

### 9.4:

```python
Z = np.random.normal(size=(n,n))
v = np.ones((n,))
def mPower(A, v, eps=1e-6, N=1e8):
    c = 1

    y = v/v[np.argmax(np.abs(v))]

    x = np.dot(A, v)
    b = x[np.argmax(np.abs(x))]
    if (np.abs(x/b - y) < eps).all():
        t = b
        return (t, y, c)
    while (np.abs(x/b - y) > eps).any() and c < N:
        c = c+1
        m = x[np.argmax(np.abs(x))]
        y = x/m
        x = A.dot(y)
        b = x[np.argmax(np.abs(x))]
    return (b, y, c)

e, vec, c= mPower(Z, v)
print("The Largest Eigen Value: ", e)
print("The Coressponding Eigen Vector: ",vec)
print("The Number of Iteratıon: ", c)
```

```python
def findNearElement(A,z):
    n,n=A.shape
    index=np.argmin(np.abs(A-z))
    return A[index//n, index%n]
```
